**Data Section**

To analyze the fires, public data from IBAMA, INPE and IBGE will be accessed, Brazilian institutes that have, among their purposes, research, preservation and environmental legislation (among others) and also from Wikipedia.

On Wikipedia we will obtain data for municipalities and population in the state of Amazonas: https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_do_Amazonas_por_popula%C3%A7%C3%A3o
List of regions and microregions: https://pt.wikipedia.org/wiki/Lista_de_mesorregi%C3%B5es_e_microrregi%C3%B5es_do_Amazonas
List of municipalities: https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_do_Amazonas

From the IBGE, data from the biome of the northern region of Brazil will be analyzed, using the txt file provided by them in the geosciences section: https://www.ibge.gov.br/geociencias/downloads-geociencias.html

Environmental fines data will be extracted from IBAMA https://dadosabertos.ibama.gov.br/dados/SICAFI/AC/Quantidade/multasDistribuidasBensTutelados.html

From INPE the data of fires recorded via satellite: http://queimadas.dgi.inpe.br/queimadas/bdqueimadas/

These data will be previously treated and modeled dimensionally to analyze the period of Aug / 2020 vs Aug / 2019
