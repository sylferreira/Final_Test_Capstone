**Intro/Business Problem**
<br>
In the last ten years, car crashes and accidents involving cars are killing human lives continuously. According to IIHS.org, only in 2018, "There were 33,654 fatal motor vehicle crashes in the United States in 2018 in which 36,560 deaths occurred. ". Based on this scenario,  this report has a point to elaborate on a predictive model based on different road conditions and weather to approach possibilities to fewer car accidents on roads. 

**Data**
This dataset has a collection of numbers and characteristics of car collisions in the US from 2004 to 2020. Was provided by the course, and has a collection of data about with date, weather conditions, road conditions, and collision descriptions.
