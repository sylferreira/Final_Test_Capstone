**Introduction & Business Problem**


**Problem Background:** In the last decade, the number of incidents in the northern region of Brazil, where the Brazilian Amazon is located, has gradually increased. From 2019 to 2020, INPE (National Institute for Space Research), there was an increase of 20% in the fires. In Brazil, the composition of the area known as the Legal Amazon is composed of all the States of Acre, Amapá, Amazonas, Mato Grosso, Pará, Rondônia, Roraima and Tocantins and part of Maranhão, these states are located in the north and northeast of the country . Of these, the largest is the state of Amazonas, the focus of this study.

**Problem Description:** The state of Amazonas, as stated earlier, is the largest state in the Northern region of Brazil. Identifying the fires in this state and observing the proximity of the microregions to the urban population proposes to make society aware of the risk to human life of deforestation and criminal fires in forest areas

**Target Audiences:** The purpose of presenting these data is to raise the awareness of voting citizens in Brazil to understand that fires in the Amazon rainforest expose an ecological biome and the lives of people in urban locations to risk.

The criterion here is to identify the municipalities in the microregions with the greatest focus of fires, in order to point out the urban centers close to the regions with outbreaks of fires
